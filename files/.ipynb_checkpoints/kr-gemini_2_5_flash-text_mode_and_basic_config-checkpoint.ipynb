```python
prompt="ì¶”ì„±í›ˆì€ ìœ ëª…í•œ ìœ ë„ì„ ìˆ˜ì´ì ì¢…í•©ê²©íˆ¬ê¸° ì„ ìˆ˜ì…ë‹ˆë‹¤. ì¶”ì„±í›ˆì€ ë¶„ë‹¹ í‰ê·  25ëŒ€ì˜ í€ì¹˜ë¥¼ ë‚ ë¦½ë‹ˆë‹¤. UFCì˜ ì¼ë°˜ ê²½ê¸° ë£°ì€ í•œ ê²½ê¸°ê°€ 5ë¶„ì”© 3 ë¼ìš´ë“œë¡œ ì§„í–‰ë©ë‹ˆë‹¤. ì¶”ì„±í›ˆì€ ì´ ëª‡ ëŒ€ì˜ í€ì¹˜ë¥¼ ë‚ ë ¸ìŠµë‹ˆê¹Œ?"
```

```python
output_text = ""
markdown_display_area = display(Markdown(output_text), display_id=True)

for chunk in client.models.generate_content_stream(
    model=MODEL_ID,
    contents=prompt,
):
    output_text += chunk.text
    output_text += "\n-----\n"  # T Kim: A separator is added to separate each chunk.
    markdown_display_area.update(Markdown(output_text))
```
